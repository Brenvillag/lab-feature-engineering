There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship_clean = spaceship.dropna()
spaceship_clean.shape
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship_clean.loc[:, 'Cabin'] = spaceship_clean['Cabin'].str.split('/').str[0]
classification = {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}
spaceship_clean.loc[:, 'Cabin'] = spaceship_clean['Cabin'].where(spaceship_clean['Cabin'].isin(classification))
spaceship_clean
```

- Drop PassengerId and Name

```python
spaceship_2 = spaceship_clean.drop(columns=["PassengerId", "Name"])
spaceship_2
```

- For non-numerical columns, do dummies.

```python
from sklearn.preprocessing import OneHotEncoder

spaceship_2.select_dtypes('object').columns
non_num_col = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
ohe= OneHotEncoder(sparse_output=False)
ohe.fit(spaceship_2[non_num_col])
```

**Perform Train Test Split**

```python
features = spaceship_2.drop(columns='Transported')
target = spaceship_2['Transported']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
X_train_trans_np = ohe.transform(X_train[non_num_col])
X_train_trans_df = pd.DataFrame(X_train_trans_np, columns=ohe.get_feature_names_out(), index=X_train.index)
X_train_trans_df
```

```python
X_train_numeric = X_train.drop(columns=non_num_col)
X_train_final = pd.concat([X_train_numeric, X_train_trans_df], axis=1)
```

```python
X_test_trans_np = ohe.transform(X_test[non_num_col])
X_test_trans_df = pd.DataFrame(X_test_trans_np, columns=ohe.get_feature_names_out(non_num_col), index=X_test.index)
X_test_num = X_test.drop(columns=non_num_col)
X_test_final = pd.concat([X_test_num, X_test_trans_df], axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train_final, y_train)
```

- Evaluate your model's performance. Comment it

```python
from sklearn.metrics import accuracy_score, classification_report
pred = knn.predict(X_test_final)
pred
y_pred = knn.predict(X_test_final)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"The accuracy of the model is {knn.score(X_test_final, y_test)*100: .2f}%")
```
